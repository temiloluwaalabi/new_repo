#  MATPLOTLIB
Plotting in matplotlib is more flexible than pandas. Instead of the dataframe or series that preceeds pandas visualization, the class plt preceeds the matplotlib visualization. There are two main ways of plotting in matplotlib; the functional method and the object oriented method.

```python
#generating random numbers
a = np.linspace(1,5,20)
b = a * 3
c = a ** 3
#print(a)
#to toggle the figure size
plt.figure(figsize = (12,6))
#the second argument in the following lines of code is used to specify the style of the line graph
plt.plot(a, 'r*', label = 'a')
plt.plot(b, 'g-.', label = 'b')
plt.plot(c, 'b--*', label = 'c')
plt.title('First matplotlib plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend();# to show tthe egends in the graph
```

```python
a = np.linspace(1,5,20)
b = a*3
c = a**3
plt.figure(figsize=(12,6))
plt.plot(a, 'r*', label = 'a')
plt.plot(b, 'g-.', label = 'b')
plt.plot(c, 'b--*', label = 'c')
plt.title('first matplotlib plot')
plt.xlabel('X axis')
plt.xlabel('Y axis')
plt.legend();
```

# MULITPLE PLOTS IN ONE FIGURE
We can use subplot() method to add more than one plot in one figure. in the image below we will use this method to separate two graphs which we will be plotted on the same axes in the previous examples.

The subplot() methods takes three arguments: they are rows, ncols and index. they indicate the number of rows, number of columns and index number of the subplots.

For instance, in our example we want to create two sub-plots in one figure such that it comes in one row and two columns and hence we pass arguments (1,2,1) and (1,2,2) in the subplot() method. Note that we have separately used title() method for both the subplots. We use suptitle() method to make centalized title for third figure

```python
plt.figure(figsize = (15,10))
plt.subplot(2,2,1)
plt.plot(a, 'g-.', label = 'A')
plt.title('Plot for A')
plt.legend()

#contros for the second axes
plt.subplot(2,1,2)
plt.plot(b, 'ro', label = 'B')
plt.title('First subplot')
plt.suptitle('First subplts');
plt.legend();
```

# OBJECT ORIENTED PLOTTING
This is the use of variables to represent classes in matplotlib. The various methods under each class can now be accessed when the variable name is called opon(inheritance).

```python
#creates the figure object
fig = plt.figure()
#adding axes to the object
ax1 = fig.add_axes([0.1,0.2,1,1])
ax2 = fig.add_axes([1.2,0.2,1,1])
# the argument([distance from the left, distance from the base, width, height])
ax1.plot(b);
ax2.plot(a);
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.5,0.5,0.4])
ax2.bar(a,c)
ax2.set_title('Small Graph')
ax1.plot(c)
ax1.set_title('Bigger Graph');
```

The method subplot() is used in the functional method while the subplots() is used in the oop method. they are both used to indicate compiler that multiple axes should be created in one figure.
While the subplot() takes the arguments to indicate the position and the index of each of the axes, the subplots takes the number of rows and columns in the graph. Unlike the subplot() method where you will need to constantly call the subplot method everytime you want to plot a new axes, the subplots(oop) just allows you index out the particular column you want to plot on


```python
fig, ax = plt.subplots(3,3, figsize = (12,9))
#this metho usually creates a tuple that contains the figure information and the axes information; that is why fig, ax is usually used to assign both values to their respectives variable names

plt.tight_layout()
#this method is used to make the various axes ir properly

#you can indes out the axes on the figure and make the graphs that are specific

#graph1
ax[1,1].plot(b, 'y>', label = 'b')
ax[1,1].set_title('Line 2, Line Plot 2')
ax[1,1].legend();

#graph2
ax[0,2].scatter(a,b, c = 'red', label = 'a,b')
ax[0,2].set_title('Line 1, Scatter Plot 3')
ax[0,2].legend();

#graph3
ax[2,0].bar(a,c, label = 'a,c')
ax[2,0].set_title('Line 3, Bar Plot 1')
ax[2,0].legend();
```
